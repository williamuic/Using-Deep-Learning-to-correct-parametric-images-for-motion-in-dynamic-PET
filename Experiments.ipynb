```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.filters import threshold_otsu
from skimage.metrics import structural_similarity as ssim_metric, peak_signal_noise_ratio as psnr_metric

def crop_image_multichannel(image, x_center, y_center, window_size=10):
    x_start = max(0, x_center - window_size)
    x_end = min(image.shape[-2], x_center + window_size)
    y_start = max(0, y_center - window_size)
    y_end = min(image.shape[-1], y_center + window_size)
    
    # Check if the image has multiple channels (1x2x256x256)
    if image.shape[1] == 2:
        return [image[0, 0, y_start:y_end, x_start:x_end], 
                image[0, 1, y_start:y_end, x_start:x_end]]
    # If single channel (1x1x256x256)
    else:
        return [image[0, 0, y_start:y_end, x_start:x_end]]

def _compute_metrics_single_channel(image1, image2):
    data_range = image1.max() - image1.min()  # We can compute this once, as it's used multiple times
    rmse = np.sqrt(np.mean((image1 - image2) ** 2))
    nmse = np.sum((image1 - image2) ** 2) / np.sum(image1 ** 2)
    nrmse_range = rmse / (np.max(image1) - np.min(image1))
    nrmse_mean = rmse / np.mean(image1)
    ssim_value, _ = ssim_metric(image1, image2, full=True, data_range=data_range)
    psnr = psnr_metric(image1, image2, data_range=data_range)
    return rmse, nmse, nrmse_range, nrmse_mean, ssim_value, psnr


def compute_metrics_separately(image1, image2):
    """
    Computes metrics only on the first channel of the given images.
    """
    return [_compute_metrics_single_channel(image1[0], image2[0])]

# Specify the x,y coordinates of the tumor's center
x_center = 87
y_center = 127

# Load the images
number_of_fake_images = 4

fake_B1_channels = crop_image_multichannel(np.load('fake_B_61_1.npy'), x_center, y_center)
fake_B2_channels = crop_image_multichannel(np.load('fake_B_61_2.npy'), x_center, y_center)
fake_B3_channels = crop_image_multichannel(np.load('fake_B_61_3.npy'), x_center, y_center)
fake_B4_channels = crop_image_multichannel(np.load('fake_B_61_4.npy'), x_center, y_center)
real_B_channels = crop_image_multichannel(np.load('real_B_61.npy'), x_center, y_center)
real_A_channels = crop_image_multichannel(np.load('real_A_61.npy'), x_center, y_center)

# Put all the fakes in a list
fake_B_list = [fake_B1_channels, fake_B2_channels, fake_B3_channels, fake_B4_channels]

# Compute metrics for real_A vs real_B once using first channel only
metrics_realA_vs_realB = compute_metrics_separately(real_A_channels, real_B_channels)
rmse, nmse, nrmse_range, nrmse_mean, ssim, psnr = metrics_realA_vs_realB[0]
print(f"Metrics for first channel of cropped_real_A vs cropped_real_B")
print(f"RMSE: {rmse}")
print(f"NMSE: {nmse}")
print(f"NRMSE (based on range): {nrmse_range}")
print(f"NRMSE (based on mean): {nrmse_mean}")
print(f"SSIM: {ssim}")
print(f"PSNR: {psnr}")
print("-" * 40)

# Compute metrics for each first channel of each fake_B vs real_B
for fake_idx, fake_B_channels in enumerate(fake_B_list, start=1):
    metrics_fake_vs_realB = compute_metrics_separately(fake_B_channels, real_B_channels)
    rmse, nmse, nrmse_range, nrmse_mean, ssim, psnr = metrics_fake_vs_realB[0]
    print(f"Metrics for first channel of cropped_fake_B_{fake_idx} vs cropped_real_B")
    print(f"RMSE: {rmse}")
    print(f"NMSE: {nmse}")
    print(f"NRMSE (based on range): {nrmse_range}")
    print(f"NRMSE (based on mean): {nrmse_mean}")
    print(f"SSIM: {ssim}")
    print(f"PSNR: {psnr}")
    print("-" * 40)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.filters import threshold_otsu
from skimage.metrics import structural_similarity as ssim_metric, peak_signal_noise_ratio as psnr_metric

def crop_image_multichannel(image, x_center, y_center, window_size=15):
    x_start = max(0, x_center - window_size)
    x_end = min(image.shape[-2], x_center + window_size)
    y_start = max(0, y_center - window_size)
    y_end = min(image.shape[-1], y_center + window_size)
    
    # Check if the image has multiple channels (1x2x256x256)
    if image.shape[1] == 2:
        return  [image[0, 1, y_start:y_end, x_start:x_end]]
    # If single channel (1x1x256x256)
    else:
        return [image[0, 0, y_start:y_end, x_start:x_end]]

def _compute_metrics_single_channel(image1, image2):
    data_range = image1.max() - image1.min()  # We can compute this once, as it's used multiple times
    rmse = np.sqrt(np.mean((image1 - image2) ** 2))
    nmse = np.sum((image1 - image2) ** 2) / np.sum(image1 ** 2)
    nrmse_range = rmse / (np.max(image1) - np.min(image1))
    nrmse_mean = rmse / np.mean(image1)
    ssim_value, _ = ssim_metric(image1, image2, full=True, data_range=data_range)
    psnr = psnr_metric(image1, image2, data_range=data_range)
    return rmse, nmse, nrmse_range, nrmse_mean, ssim_value, psnr


def compute_metrics_separately(image1, image2):
    """
    Computes metrics only on the first channel of the given images.
    """
    return [_compute_metrics_single_channel(image1[0], image2[0])]

# Specify the x,y coordinates of the tumor's center
x_center = 87
y_center = 127

# Load the images
number_of_fake_images = 4

fake_B1_channels = crop_image_multichannel(np.load('fake_B_61_1.npy'), x_center, y_center)
fake_B2_channels = crop_image_multichannel(np.load('fake_B_61_2.npy'), x_center, y_center)
fake_B3_channels = crop_image_multichannel(np.load('fake_B_61_3.npy'), x_center, y_center)
fake_B4_channels = crop_image_multichannel(np.load('fake_B_61_4.npy'), x_center, y_center)
real_B_channels = crop_image_multichannel(np.load('real_B_61.npy'), x_center, y_center)
real_A_channels = crop_image_multichannel(np.load('real_A_61.npy'), x_center, y_center)

# Put all the fakes in a list
fake_B_list = [fake_B1_channels, fake_B2_channels, fake_B3_channels, fake_B4_channels]

# Compute metrics for real_A vs real_B once using first channel only
metrics_realA_vs_realB = compute_metrics_separately(real_A_channels, real_B_channels)
rmse, nmse, nrmse_range, nrmse_mean, ssim, psnr = metrics_realA_vs_realB[0]
print(f"Metrics for first channel of cropped_real_A vs cropped_real_B")
print(f"RMSE: {rmse}")
print(f"NMSE: {nmse}")
print(f"NRMSE (based on range): {nrmse_range}")
print(f"NRMSE (based on mean): {nrmse_mean}")
print(f"SSIM: {ssim}")
print(f"PSNR: {psnr}")
print("-" * 40)

# Compute metrics for each first channel of each fake_B vs real_B
for fake_idx, fake_B_channels in enumerate(fake_B_list, start=1):
    metrics_fake_vs_realB = compute_metrics_separately(fake_B_channels, real_B_channels)
    rmse, nmse, nrmse_range, nrmse_mean, ssim, psnr = metrics_fake_vs_realB[0]
    print(f"Metrics for first channel of cropped_fake_B_{fake_idx} vs cropped_real_B")
    print(f"RMSE: {rmse}")
    print(f"NMSE: {nmse}")
    print(f"NRMSE (based on range): {nrmse_range}")
    print(f"NRMSE (based on mean): {nrmse_mean}")
    print(f"SSIM: {ssim}")
    print(f"PSNR: {psnr}")
    print("-" * 40)
```

```python
def visualize_all_in_row(images, x_center, y_center, colorbar_range=None, window_size=15):
    # Crop the images
    cropped_images = [crop_image_multichannel(img, x_center, y_center, window_size)[0] for img in images]  # Selecting only the first channel for 1x2x256x256
    
    fig, axes = plt.subplots(1, len(cropped_images), figsize=(3 * len(cropped_images), 5))
    
    # If no colorbar_range is given, set it based on the second image (assumed to be real_B here)
    if colorbar_range is None:
        colorbar_range = (cropped_images[1].min(), cropped_images[1].max())
    
    for ax, image,  in zip(axes, cropped_images):
        im = ax.imshow(image, vmin=colorbar_range[0], vmax=colorbar_range[1])
        ax.axis('off')

    # Add colorbar on the right side without adjusting the image plots
    plt.subplots_adjust(wspace=0.01)
    plt.tight_layout()
    plt.show()

# Load the images
fake_B_files = ["fake_B_61_1.npy", "fake_B_61_2.npy", "fake_B_61_3.npy","fake_B_61_4.npy"]  
fake_B_images = [np.load(path) for path in fake_B_files]

real_A = np.load('real_A_61.npy')
real_B = np.load('real_B_61.npy')

# Specify the x,y coordinates of the tumor's center
x_center = 87
y_center = 127

# Combine all the images in a list in the desired order and visualize
all_images = [real_A, real_B] + fake_B_images

visualize_all_in_row(all_images, x_center, y_center)
```
